### Simplest Neural Network

```python
from graphviz import Digraph

# Create a directed graph
dot = Digraph()

# Set the layout to left-to-right
dot.attr(rankdir='LR')

# Add nodes for each layer
dot.node('A', 'Input X')
dot.node('B', 'W1(x) + b1')
dot.node('C', 'W2(W1(x)+b1) + b2')

# Connect nodes to show forward pass flow
dot.edges(['AB', 'BC'])

# Display the graph
dot
```
